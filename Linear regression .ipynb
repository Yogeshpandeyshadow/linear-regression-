```python
import statsmodels.api as sm
```

```python
filepath = "C:/Users/DELL/OneDrive/Desktop/IIT DHANBAD SAMRAT/7. Multivariate Lab/Practice/Exp 6 Data.csv"
df=pd.read_csv(filepath)
df
```

```python
x=df.iloc[:,2].copy()
```

```python
x
```

```python
y=df['y'].copy()
```

```python
x,y = np.array(x),np.array(y)
x=sm.add_constant(x)
model1=sm.OLS(y,x)
results1=model1.fit()
print(results1.summary())
print("\n Fitted Values: \n ")
y_pred=results1.fittedvalues.round(2)
print(y_pred)
```
